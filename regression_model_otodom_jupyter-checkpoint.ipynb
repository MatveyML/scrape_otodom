```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
df = pd.read_csv(r"C:\Users\User\Documents\Python\otodom\clean_otodom.csv")
pd.set_option("display.max_columns", None)
# categorical to numerical
df = pd.get_dummies(df, columns=['District','Owner','Building'])
# remove the outliers that are more than 3 standard deviations away from the mean
df['price_z'] = np.abs(stats.zscore(df['Full_price']))
df = df[df['price_z'] <= 3]
df = df[df['Full_price'] != 0]
df = df.drop('price_z', axis=1)
plt.hist(df['Full_price'], bins=100)
```

```python
y = np.asanyarray(df['Full_price'])
x = np.asanyarray(df.drop('Full_price',axis=1))
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
#divide into sets
x_train_scaled, x_test_scaled, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
ridge = Ridge(alpha=0.1)
ridge.fit(x_train_scaled,y_train)
yhat_ridge = ridge.predict(x_test_scaled)
r2_ridge = r2_score(y_test,yhat_ridge)
print('R^2 for Ridge regression:',r2_ridge)
```

```python
residuals = y_test - yhat_ridge
plt.scatter(yhat_ridge, residuals, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='-')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residual Plot for Ridge regression')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
y_reshaped = y.reshape(-1,1).ravel()
x_scaled = scaler.fit_transform(x)
x_train_scaled, x_test_scaled, y_train_reshaped, y_test_reshaped = train_test_split(x_scaled, y_reshaped, test_size=0.2, random_state=42)
rf_model = RandomForestRegressor(n_estimators=150, random_state=42)
rf_model.fit(x_train_scaled,y_train_reshaped)
yhat_rf = rf_model.predict(x_test_scaled)
r2_rf = r2_score(y_test_reshaped,yhat_rf)
print('R^2 for rf:',r2_rf)
```
